# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
cur_x = 3 # The algorithm starts at x=3
precision = 0.000001 #This tells us when to stop the algorithm
max_iters = 10000 # maximum number of iterations
df = lambda x: 2*(x+5) #Gradient of our function 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
'''
需要執行初始值設定, 下列三個
# Learning rate
#iteration counter
#Step size  
'''
'''
while previous_step_size > precision and iters < max_iters:

  算法迭代更新

print("the learning rate is",lr, "\nIteration",iters,"\nX value is",cur_x) #Print iterations
'''
def GradientDescent(learning_rate):
    cur_x = 3 # The algorithm starts at x=3
    precision = 0.000001 #This tells us when to stop the algorithm
    max_iters = 10000 # maximum number of iterations
    df = lambda x: 2*(x+5) #Gradient of our function 
    iteration_counter = 0 
    Step_size = 0.1
    lr = learning_rate
    # Record of each iteration
    iters_history = [iteration_counter]
    x_history = [cur_x]
    
    while Step_size > precision and iteration_counter < max_iters: # 當x移動步伐<precision 或 和回數超過預設 則停止
        prev_x = cur_x #Store current x value in prev_x
        cur_x = cur_x - lr * df(prev_x) #Gradient descent
        Step_size = abs(cur_x - prev_x) # 取較大的值, Change in x # Reset前面x移動的步伐
        iteration_counter = iteration_counter+1 #iteration count
        #print("Iteration",iteration_counter,"\nX value is",cur_x) #Print iterations
         # Store parameters for plotting
        iters_history.append(iteration_counter) # add new iteration i
        x_history.append(cur_x) # add new x i

    print("Totally iteations: ", iteration_counter)
    print("The local minimum occurs at", cur_x)
    # plot results
    plt.plot(iters_history, x_history, 'o-', ms=3, lw=1.5, color='black') # x_axis= iteration/ y_axis= position of x
    plt.xlabel(r'$iters$', fontsize=16)
    plt.ylabel(r'$x$', fontsize=16)
    plt.show()
```

```python
learning_rate = [0.1, 0.0001]
print('learning_rate=0.1=================================')
GradientDescent(learning_rate[0])
```

```python
print('learning_rate=0.0001==============================')
GradientDescent(learning_rate[1])
```
